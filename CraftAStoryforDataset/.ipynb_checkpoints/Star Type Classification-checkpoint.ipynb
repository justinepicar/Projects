# CLASSIFYING STARS

Using the Star Data Classification/NASA dataset from Kaggle (source: https://www.kaggle.com/brsdincer/star-type-classification), we can explore which features are significant in classifying stars.

This can be useful for budding astronomers who want to be more knowledgable about what characteristics are most important when observing stars or helping astrophysicists label new stars with ease. 

TARGET:

Type from 0 to 5:

- Red Dwarf - 0
- Brown Dwarf - 1
- White Dwarf - 2
- Main Sequence - 3
- Super Giants - 4
- Hyper Giants - 5

FEATURES:

- Temperature
- Luminosity -- Relative Luminosity
- Radius -- Relative Radius
- Abs_Magnitude -- Absolute Magnitude
- Color -- General Color of Spectrum
- Spectral_Class -- O,B,A,F,G,K,M / SMASS - https://en.wikipedia.org/wiki/Asteroid_spectral_types


MATH:

- Lo = 3.828 x 10^26 Watts
- (Avg Luminosity of Sun)
- Ro = 6.9551 x 10^8 m
- (Avg Radius of Sun)

```python
#import packages to use
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#download and read csv file
df=pd.read_csv(r'Stars.csv', header=0, 
               names=['Temperature', 'Luminosity', 'Radius', 'Abs_Magnitude', 'Color', 'Spectral_Class', 'Type'])
```

```python
df.head()
```

A quick look at the data below shows that Type needs to be changed to a string type since it represents a categorical variable. There are no missing data points. 

```python
df.info()
```

```python
df.hist(figsize=(15, 15))
plt.subplots_adjust(hspace=0.5)
```

```python
df.Type = df.Type.astype(str)
```

```python
numerics = ['int64','float64']
for c in [c for c in df.columns if df[c].dtype in numerics]:
    df[c] = np.log(df[c])
```

```python
df.hist(figsize=(15, 15))
plt.subplots_adjust(hspace=0.5)
```

First, we'll use a heatmap to see any correlations between the log numerical features of the data.

```python
sns.heatmap(data=df.corr(), annot=True)
plt.title('Pearson Correlation of Numerical Features')
plt.show()
```

Right away we can see that there is a very strong negative correlation between Luminosity and Absolute Magnitude.
There is a very strong positive correlation between Luminosity and Radius. This suggests that bigger stars emit a larger amount of energy or wattage from its surface and that more luminous an object, the smaller the absolute magnitude.

```python
plt.figure(figsize=(15,15))
sns.relplot(x='Abs_Magnitude', y='Luminosity', hue='Type', data=df, kind='scatter')
plt.title('Correlation Between Abs_Magnitude and Luminosity by Type')
plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.relplot(x='Radius', y='Luminosity', hue='Type', data=df, kind='scatter')
plt.title('Correlation Between Radius and Luminosity by Type')
plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.relplot(x='Abs_Magnitude', y='Radius', hue='Type', data=df, kind='scatter')
plt.title('Correlation Between Absolute Magnitude and Radius')
plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.relplot(x='Temperature', y='Luminosity', hue='Type', data=df, kind='scatter')
plt.title('Correlation Between Temperature and Luminosity')
plt.show()
```

Based on observations above, we can see that each Type is clustered distinctly. This is especially noted by Luminosity, Absolute Magnitude, and the Radius of the star. Discerning star type by temperature isn't as effective since the Types have a much larger distribution as shown above and below.

```python
plt.figure(figsize=(10,10))
sns.catplot(x='Temperature', y='Type', data=df, kind='violin')
plt.show()
```

Based on the figures below, we can confirm that Luminosity and Radius are potentially better indicators of star type as the distributions are more distinct for each type.

```python
plt.figure(figsize=(10,10))
sns.catplot(x='Luminosity', y='Type', data=df, kind='violin')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.catplot(x='Radius', y='Type', data=df, kind='violin')
plt.show()
```

A look at the categorical features below shows that Type 0 and 1 can belong to at most 1 color and spectral class.
Type 2 and 3 varies the most in color. Type 5 varies the most by spectral class.

```python
plt.figure(figsize=(10,10))
countplt = sns.catplot(x='Color', col='Type', col_wrap=3, data=df, kind='count')
countplt.set_xticklabels(rotation=90)
plt.suptitle('Types of Stars by Color')
plt.subplots_adjust(top=0.88)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.catplot(x='Spectral_Class', col='Type', col_wrap=2, data=df, kind='count')
plt.suptitle('Types of Stars by Spectral_Class')
plt.subplots_adjust(top=0.95)
plt.show()
```

After taking a cursory look at the data above, we can come to the conclusion that the size and the luminosity of the star may be the most useful features to predict star type.
